```python
import numpy as np
import os
import torch
import h5py
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter


scene_id = "0000"
data_path = "/rhome/jhou/data/dataset/megadepth/"
min_overlap_ratio = 0.3
max_overlap_ratio = 1.0

scene_info = np.load(os.path.join(data_path, 'scene_info', scene_id + '.npz'), allow_pickle=True)
overlap_matrix = scene_info['overlap_matrix']

valid =  np.logical_and(
        overlap_matrix >= min_overlap_ratio,
        overlap_matrix <= max_overlap_ratio)

image_paths = scene_info['image_paths']
depth_paths = scene_info['depth_paths']
intrinsics = scene_info['intrinsics']
poses = scene_info['poses']   

pairs = np.vstack(np.where(valid))
                

     
```

```python
def grid_positions(h, w, device):
    lines = torch.arange(0, h, device=device).view(-1, 1).float().repeat(1, w)
    columns = torch.arange(0, w, device=device).view(1, -1).float().repeat(h, 1)
    return torch.cat([lines.view(1, -1), columns.view(1, -1)], dim=0)

def uv_to_pos(uv):
    return torch.cat([uv[1, :].view(1, -1), uv[0, :].view(1, -1)], dim=0)
```

```python
def warp(depth1, intrinsics1, pose1, depth2, intrinsics2, pose2):

    # meshgrid
    device = depth1.device
    pos = grid_positions(depth1.shape[0], depth1.shape[1], device)
    X1 = pos[1,:]
    Y1 = pos[0,:]
    Z1 = depth1[Y1.long(), X1.long()].float()

    XYZ1_hom = torch.cat([
        ((X1 - intrinsics1[0, 2]) * (Z1 / intrinsics1[0, 0])).view(1, -1),
        ((Y1 - intrinsics1[1, 2]) * (Z1 / intrinsics1[1, 1])).view(1, -1),
        Z1.view(1, -1),
        torch.ones(1, Z1.size(0), device=device)
    ], dim=0)
    XYZ2_hom = torch.chain_matmul(pose2, torch.inverse(pose1), XYZ1_hom)
    XYZ2 = XYZ2_hom[: -1, :] / XYZ2_hom[-1, :].view(1, -1)

    XYZ2 = torch.matmul(intrinsics2, XYZ2)

    X2 = XYZ2[0, :] / XYZ2[2, :]
    Y2 = XYZ2[1, :] / XYZ2[2, :]
    Z2 = XYZ2[2, :]
    h_mask = (Y2 < depth2.shape[0]) & (Y2 >= 0)
    w_mask = (X2 < depth2.shape[1]) & (X2 >= 0)
    spatial_mask = h_mask & w_mask
    X2 = X2[spatial_mask].long()
    Y2 = Y2[spatial_mask].long()
    Z2 = Z2[spatial_mask]
    depth2 = depth2[Y2,X2]
    depth_mask = torch.abs(depth2 - Z2) < 0.05
    print(X2.shape)

    return torch.stack([X1,Y1],1)[spatial_mask][depth_mask], torch.stack([X2,Y2], 1)[depth_mask]
```

```python
def plot_matches(im1, im2, matches, radius=5, dpi=150, savefig=False):
    import matplotlib.pyplot as plt
    I1 = Image.fromarray(im1)
    I2 = Image.fromarray(im2)    
        
    w1, h1 = I1.size
    w2, h2 = I2.size 

    if h1 <= h2:
        scale1 = 1
        scale2 = h1/h2
        w2 = int(scale2 * w2)
        I2 = I2.resize((w2, h1))
    else:
        scale1 = h2/h1
        scale2 = 1
        w1 = int(scale1 * w1)
        I1 = I1.resize((w1, h2))
    catI = np.concatenate([np.array(I1), np.array(I2)], axis=1)

    # Load all matches
    match_num = matches.shape[0]
    print('Plotting inliers: ', match_num)

    x1 = scale1*matches[:, 0]
    y1 = scale1*matches[:, 1]
    x2 = scale2*matches[:, 2] + w1
    y2 = scale2*matches[:, 3]
    c = np.random.rand(match_num, 3) 
    
    # Plot images and matches
    fig = plt.figure(figsize=(30, 20))
    axis = plt.gca()#fig.add_subplot(1, 1, 1)
    axis.imshow(catI)
    axis.axis('off')
    
    for i in range(match_num):
        # Plot
        axis.add_artist(plt.Circle((x1[i], y1[i]), radius=radius, color=c[i,:]))
        axis.add_artist(plt.Circle((x2[i], y2[i]), radius=radius, color=c[i,:]))
        # draw lines
        axis.plot([x1[i], x2[i]], [y1[i], y2[i]], c=c[i,:], linestyle='-', linewidth=radius)
    if savefig:        
        fig.savefig(savefig, dpi=dpi,  bbox_inches='tight')      
    plt.show()    
```

```python

counter = 0

for pair_idx in range(pairs.shape[1]):
    counter += 1
    idx1 = pairs[0, pair_idx]
    idx2 = pairs[1, pair_idx]
    pair_metadata = {
        'image_path1': image_paths[idx1],
        'depth_path1': depth_paths[idx1],
        'intrinsics1': intrinsics[idx1],
        'pose1': poses[idx1],
        'image_path2': image_paths[idx2],
        'depth_path2': depth_paths[idx2],
        'intrinsics2': intrinsics[idx2],
        'pose2': poses[idx2]}

    depth_path1 = os.path.join(data_path, pair_metadata['depth_path1'])
    with h5py.File(depth_path1, 'r') as hdf5_file:
        depth1 = np.array(hdf5_file['/depth'])
    image_path1 = os.path.join(data_path, pair_metadata['image_path1'])
    image1 = Image.open(image_path1)
    #w1,h1= image1.size
    image1 = np.array(image1)
    #plt.imshow(image1)
    #plt.show()

    intrinsics1 = pair_metadata['intrinsics1']
    pose1 = pair_metadata['pose1']

    depth_path2 = os.path.join(data_path, pair_metadata['depth_path2'])
    with h5py.File(depth_path2, 'r') as hdf5_file:
        depth2 = np.array(hdf5_file['/depth'])
    image_path2 = os.path.join(data_path, pair_metadata['image_path2'])
    image2 = Image.open(image_path2)
    image2 = np.array(image2)

    #plt.imshow(image2)
    #plt.show()

    intrinsics2 = pair_metadata['intrinsics2']
    pose2 = pair_metadata['pose2']

    depth1 = torch.from_numpy(depth1).float()
    intrinsics1 = torch.from_numpy(intrinsics1).float()  # [3, 3]
    pose1 = torch.from_numpy(pose1).float()


    depth2 = torch.from_numpy(depth2).float()
    intrinsics2 = torch.from_numpy(intrinsics2).float()  # [3, 3]
    pose2 = torch.from_numpy(pose2).float()


    # Warp the positions from image 1 to image 2
    inds1, inds2 = warp(depth1, intrinsics1, pose1, depth2, intrinsics2, pose2)

    matches = torch.cat([inds1[0:100000:10000],inds2[0:100000:10000]],1)
    #draw matching
    # matches
    plot_matches(image1, image2, matches)
    break
    

    


```
